Importing the Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

Data Collection and Processing


```python
# loading the csv data to a Pandas DataFrame
gold_data = pd.read_csv('/content/gold_price_data.csv')
```

Correlation:

```python
correlation = gold_data.select_dtypes(include=['number']).corr()
```

```python
# constructing a heatmap to understand the correlatiom
plt.figure(figsize = (8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the Features and Target


```python
X = gold_data.drop(['Date','GLD'],axis=1)
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

Splitting into Training data and Test Data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

Model Training:
Random Forest Regressor


```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
# training the model
regressor.fit(X_train,Y_train)
```

Model Evaluation


```python
# prediction on Test Data
test_data_prediction = regressor.predict(X_test)
```

```python
print(test_data_prediction)
```

Model results


```python
def error_details():
    # R squared error
    r2_error = metrics.r2_score(Y_test, test_data_prediction)

    # Mean Absolute Error (MAE)
    mae_error = metrics.mean_absolute_error(Y_test, test_data_prediction)

    # Mean Squared Error (MSE)
    mse_error = metrics.mean_squared_error(Y_test, test_data_prediction)

    # Root Mean Squared Error (RMSE)
    rmse_error = np.sqrt(mse_error)

    # Print all errors
    print('R squared error: ', r2_error)
    print('Mean Absolute Error (MAE): ', mae_error)
    print('Mean Squared Error (MSE): ', mse_error)
    print('Root Mean Squared Error (RMSE): ', rmse_error)
```
